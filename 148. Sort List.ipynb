```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def sortList(self, head: ListNode) -> ListNode:
        if not head:
            return(head);
        #print(head);
        p=head;
        l=0;
        while p:
            p=p.next;
            l+=1;
        slide=1;
        f_head=ListNode(0);
        f_head.next=head;
        while slide<l:
            flag=f_head;
            while flag.next:
                p1_head=flag.next;
                flag.next=None
                p1_tail=self.after(p1_head,slide-1);
                if p1_tail and p1_tail.next:
                    p2_head=p1_tail.next;
                    p1_tail.next=None;
                    p2_tail=self.after(p2_head,slide-1);
                    tmp=p2_tail.next if p2_tail else None;
                    if p2_tail: p2_tail.next=None;
                else:
                    p2_head=None;
                    p2_tail=None;
                    tmp=None;
                #print(p1_head);
                #print(p2_head);
                p_head,p_tail=self.merge(p1_head,p2_head);
                #print(p_head)
                flag.next=p_head;
                p_tail.next=tmp;
                flag=p_tail;
            slide*=2;
        return(f_head.next)
    def after(self,p:ListNode,l):
        if not p:
            return(None);
        for i in range(l):
            if p.next: 
                p=p.next;
            else:
                return(None);
        return(p);
    def merge(self,p1:ListNode,p2:ListNode):
        head=ListNode(0);
        p=head;   
        while p1 and p2:
            if p1.val<p2.val:
                p.next=p1;
                p=p.next;
                p1=p1.next;
                p.next=None;
            else:
                p.next=p2;
                p=p.next;
                p2=p2.next;
                p.next=None;
        if p1:
            p.next=p1;
        else:
            p.next=p2;
        while p.next:
            p=p.next;
        return(head.next,p);
```
